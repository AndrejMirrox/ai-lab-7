# Индивидуальное задание

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_gray = cv2.imread('pic/Ind.jpg', 0)
image_color = cv2.imread('pic/Ind.jpg', 1)
```

```python
im_bw = cv2.threshold(image_color, 128, 255, cv2.THRESH_BINARY_INV)[1]
im_bwa = cv2.threshold(image_color, 128, 255, cv2.THRESH_BINARY)[1]
im_bwb = cv2.threshold(image_gray, 128, 255, cv2.THRESH_BINARY_INV)[1]
```

```python
plt.subplot(2, 3, 1)
plt.imshow(image_gray)
plt.axis("off")
plt.subplot(2, 3, 2)
plt.imshow(image_color)
plt.axis("off")
plt.subplot(2, 3, 3)
plt.imshow(im_bw)
plt.axis("off")
plt.subplot(2, 3, 4)
plt.imshow(im_bwa)
plt.axis("off")
plt.subplot(2, 3, 5)
plt.imshow(im_bwb)
plt.axis("off");
```

```python
image = cv2.rectangle(image_color, (230, 180), (360, 420), (255, 0, 255), 10)
plt.imshow(image);
```
